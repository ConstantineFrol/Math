# Тема “Пределы”


1) На языке Python предложить алгоритм вычисляющий численно предел с точность e = 10^(-7)


$$ \lim_{n \to \infty} \frac{n}{\sqrt[n]{n!} \qquad } $$

```python
import math
from sys import setrecursionlimit
```

```python
setrecursionlimit(10 ** 9)
accuracy = 10 ** -7
a = 0
a_n = 1
n = 1

while abs(a_n - a) >= accuracy:
    a = n / (math.factorial(n) ** (1 / n))
    n += 1
    a_n = n / (math.factorial(n) ** (1 / n))
    print(abs(a_n - a))
```

#### 1) Задание
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10). 
* Найдите их сумму. (на листочке)
* Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
A = np.array([10, 10, 10])
B = np.array([0, 0, -10])

# Сумма
AB_sum = A + B
print("Cумма AB:", AB_sum)

# Найдем вертор по координатам
AB = B - A
AB_length = np.sqrt(np.sum(np.power(AB, 2))) # |AB| = (ABx^2 + ABy^2 + ABz^2)^1/2
print("Длина вектора:", AB_length)
```

#### 3. Задание (в программе)  Напишите код на Python, реализующий построение графиков:
* окружности,	
* эллипса,
* гиперболы.


```python
x1 = []
y1 = []
x2 = []
y2 = []
for i in range(1000):
    r = 50
    x1.append(i)
    x2.append(-i)
    y1.append(np.sqrt(r**2 - i**2))
    y2.append(-np.sqrt(r**2 - i**2))
    
plt.plot(x1, y1, color='r')
plt.plot(x1, y2, color='r')
plt.plot(x2, y1, color='r')
plt.plot(x2, y2, color='r')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Окружность")
plt.axis('scaled')
plt.show()
```

```python
x1 = []
x2 = []
y1 = []
y2 = []

for i in range(1000):
    a = 130
    b = 90
    x1.append(-i)
    x2.append(i)
    y1.append(np.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))
    y2.append(-np.sqrt(b ** 2 - (b ** 2 * (i ** 2 / a ** 2))))
       
plt.plot(x1, y1, color='r')
plt.plot(x1, y2, color='r')
plt.plot(x2, y2, color='r')
plt.plot(x2, y1, color='r')
plt.axis('scaled')
plt.title("Элипс")
plt.show()
```

```python
x1 = []
x2 = []
y1 = []
y2 = []

for i in range(1000):
    a = 200
    b = 100
    x1.append(-i)
    x2.append(i)
    y1.append(np.sqrt(((i*b)/a)**2 - b**2))
    y2.append(-np.sqrt(((i*b)/a)**2 - b**2))
       
plt.plot(x1, y1, color='r')
plt.plot(x1, y2, color='r')
plt.plot(x2, y2, color='r')
plt.plot(x2, y1, color='r')
plt.axis('scaled')
plt.title("Гипербола")
plt.show()
```
